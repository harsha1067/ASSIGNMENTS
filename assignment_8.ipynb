```python
import pandas as pd 
import numpy as np
import seaborn as sns 
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize 
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('wine.csv')
df
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
df_ary=df.values 
df_ary
```

```python
df_norm=scale(df_ary)
df_norm
```

```python
df1=PCA()
df1_values=df1.fit_transform(df_norm)
df1_values
```

```python
df1.components_
```

```python
var=df1.explained_variance_ratio_ 
var
```

```python
df1_values[:,0:1]
```

```python
x=df1_values[:,0:1]
y=df1_values[:,1:2]
plt.scatter(x,y);
```

```python
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
plt.plot(var1,color='pink')
```

```python
df_f=pd.concat([df['Type'],pd.DataFrame(df1_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
df_f
```

```python
fig=plt.figure(figsize=(22,12))
sns.scatterplot(data=df_f)
```

```python
sns.scatterplot(data=df_f, x='PC1', y='PC2', hue='Type');
```

```python
plt.figure(figsize=(12,6))
dendrogram = sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
import warnings
warnings.filterwarnings('ignore')
x= pd.DataFrame(hclusters.fit_predict(df_norm),columns=['clustersid'])
x['clustersid'].value_counts()
```

```python
df3=df.copy()
df3['clustersid']=hclusters.labels_
df3
```

```python
daf=[]
for i in range(6,12):
  kmeans=KMeans(n_clusters=i,random_state=3)
  kmeans.fit(df_norm)
  daf.append(kmeans.inertia_)
```

```python
plt.plot(range(6,12),daf)
plt.title('CLUSTER')
```

```python
daf_c=KMeans(6,random_state=22).fit(df_norm)
daf_c
```

```python
df3['daf_cid'].value_counts() 
```

```python
daf_c.labels_
```

```python
df3=df.copy()
df3['daf_cid']=daf_c.labels_
df3
```
