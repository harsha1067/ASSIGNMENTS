```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import string
import spacy 
from matplotlib.pyplot import imread 
from wordcloud import WordCloud, STOPWORDS 
%matplotlib inline
import re
```

```python
df = pd.read_csv("Elon_musk.csv", encoding = 'latin-1')
df
```

```python
df.drop(['Unnamed: 0'], inplace = True, axis = 1)
df
```

```python
df = [Text.strip() for Text in df.Text]
df = [Text for Text in df if Text]
df[0:11]
```

```python
df_text = ' '.join(df)
df_text
```

```python
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer(strip_handles = True)
df_tokens = tknzr.tokenize(df_text)
print(df_tokens)
```

```python
df_tokens_text = ' '.join(df_tokens)
df_tokens_text 
```

```python
df_no_punc_text = df_tokens_text.translate(str.maketrans(' ', ' ', string.punctuation))
df_no_punc_text 
```

```python
no_url_text = re.sub('', '', df_no_punc_text)
no_url_text
```

```python
from nltk.tokenize import word_tokenize 
text_tokens = word_tokenize(no_url_text)
print(text_tokens) 
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
len(text_tokens)
```

```python
from nltk.corpus import stopwords 
my_stop_words = stopwords.words('english')
list1 = ['\*92', 'rt', 'ye', 'yeah', 'Yes', 'u0001F923', 'I']
my_stop_words.extend(list1)
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```python
lower_words = [Text.lower() for Text in no_stop_tokens]
print(lower_words[100:200])
```

```python
from nltk.stem import PorterStemmer 
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[100:200])
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp(' '.join(lower_words))
print(doc)
```

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas)
```

```python
c_tweets = ' '.join(lemmas)
c_tweets
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer()
tweetscv = cv.fit_transform(lemmas)
print(cv.vocabulary_ )
```

```python
print(cv.get_feature_names_out()[100:200])
```

```python
print(tweetscv.toarray()[100:200])
```

```python
cv_ngram_range = CountVectorizer(analyzer = 'word', ngram_range = (1,3), max_features = 100)
bow_matrix_ngram = cv_ngram_range.fit_transform(lemmas)
```

```python
print(cv_ngram_range.get_feature_names_out())
print(bow_matrix_ngram.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidfv_ngram_max_features = TfidfVectorizer(norm = 'l2', analyzer = 'word', ngram_range = (1,3), max_features = 600)
tfidf_matix_ngram = tfidfv_ngram_max_features.fit_transform(lemmas)
```

```python
print(tfidfv_ngram_max_features.get_feature_names_out())
print(tfidf_matix_ngram.toarray())
```

```python
def plot_cloud(wordcloud):
  plt.figure(figsize = (40,20))
  plt.imshow(wordcloud)
  plt.axis('off')
STOPWORDS.add('pron')
STOPWORDS.add('rt')
STOPWORDS.add('yeah')
wordcloud = WordCloud(width = 1500, height = 1000, background_color = 'black', max_words = 60, colormap = 'Set1', stopwords = STOPWORDS).generate(c_tweets)
plot_cloud(wordcloud) 
```

```python
import warnings
warnings.filterwarnings('ignore')
nlp = spacy.load('en_core_web_sm')
one_block = c_tweets
doc_block = nlp(one_block)
spacy.displacy.render(doc_block, style = 'ent', jupyter = True)
```

```python
for token in doc_block[100:200]:
  print(token, token.pos_)
```

```python
nouns_verbs = [df.text for df in doc_block if df.pos_ in ('NOUN', 'VERB')]
print(nouns_verbs[100:200])
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer()
X = cv.fit_transform(nouns_verbs)
sum_words = X.sum(axis = 0)
words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
word_freq = sorted(words_freq, key = lambda x : x[1], reverse = True)
w_df = pd.DataFrame(words_freq)
w_df.columns = ['word', 'count']
w_df[0:11]
```

```python
w_df[0:6].plot.bar(x = 'word', figsize = (12,6), title = "NOUNS & VERBS")
```
