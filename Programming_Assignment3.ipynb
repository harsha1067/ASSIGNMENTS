```python
#Create 5 dataframes(eng,state,party,liberal,ivies) from the 5 csv files uploaded above.
eng_url = pd.read_csv(eng_url)
state_url = pd.read_csv(state_url)
party_url = pd.read_csv(party_url)
liberal_url = pd.read_csv(liberal_url)
ivies_url = pd.read_csv(ivies_url)
```

```python
#Use a function to retreive information on the first 10 rows of your dataframe.
def get_first_10_rows(dataframe):
  '''
  Retrieve info about the first 10 rows of a dataframe.

  Parameters:
  - dataframe: PandasDataFrame

  Returns:
  - None (prints information)
  '''
  print(f"Shape of the dataframe: {dataframe.shape}")
  print("\nFirst 10 rows of the dataframe:")
  print(dataframe.head(10))
  print("\nSummary statistics:")
  print(dataframe.describe())
  print("\nInfo about the dataframe:")
  print(dataframe.info())

get_first_10_rows(eng_url)
get_first_10_rows(state_url)
get_first_10_rows(party_url)
get_first_10_rows(liberal_url)
get_first_10_rows(ivies_url)
```
