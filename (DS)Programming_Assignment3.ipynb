```python
#Dataframe creation
games_url = 'https://andybek.com/pandas-games'
```

```python
games_df = pd.read_csv(games_url)
games_df.head(11)
```

Q7: What are the total sales across all regions?

```python
#Q7-Code
total_sales = games_df['Global_Sales'].sum()
print(f"Total sales of all regions: {total_sales}")
```

 Q8: Find the total sales by region for X360 and PS3.


```python
#8-Code
filtered_games = games_df[games_df['Platform'].isin(['X360', 'PS3'])]    # Filter the dataframe for 'X360' and 'PS3' platforms

total_sales_by_reg = filtered_games.groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum()      # Calculate total sales by region
print("Total sales by region for X360 and PS3:")
total_sales_by_reg
```

Q9: How much does each platform sell across regions, on average?

```python
#Q9-Code
average_sales = games_df.groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].mean()     # Calculating average sales for each platform across regions

# Display the result
print("Average Sales by Platform Across Regions:")
average_sales
```

Q10: Create a smaller dataframe from *games*, selecting only the Publisher, Genre, Platform, and NA_Sales columns. Assign this dataframe to the variable *publishers*.

```python
#Q10-Code
publishers = games_df[['Publisher', 'Genre', 'Platform', 'NA_Sales']]
publishers.head(11)
```

Q11: From the *publishers* dataframe, find the top 10 game publishers in North America by total sales.

```python
#Q11-Code
top_publishers_NA = publishers.groupby('Publisher')['NA_Sales'].sum()     #Group by 'Publisher' and calculate total sales in North America

top_10 = top_publishers_NA.sort_values(ascending=False).head(10)       #Sorting in descending order and get the top 10 publishers

print("Top 10 Game Publishers in North America by Total Sales:")
print(top_10)
```

Q12: What is the gaming platform that has attracted the most sales in North America?



```python
#Q12:Code
platform_NA = publishers.groupby('Platform')['NA_Sales'].sum()      #Group by 'Platform' and calculate total sales in North America

# Find the platform with the most sales in North America
most_sales = platform_NA.idxmax()

# Display the result
print("Gaming Platform with the Most Sales in North America:", most_sales)
```

```python
#Create Dataframe sat
sat_url = 'https://andybek.com/pandas-sat'

sat = pd.read_csv(sat_url)
sat
```

Q13: Starting with our main dataframe (*sat*), create a pivot table that summarizes the <ins>total</ins> student enrollment (Student Enrollment) across all 5 boroughs (*Borough*).

Which boroughs have the highest and lowest high school student bodies?

```python
#Q13
pivot_table_ent = sat.pivot_table(values = 'Student Enrollment', index = 'Borough', aggfunc = 'sum')       #Pivot table to summarize total student enrollment across all 5 boroughs

highest_enrollment_borough = pivot_table_ent.idxmax()
lowest_enrollment_borough = pivot_table_ent.idxmin()

print(f"Borough with the Highest Total Student Enrollment: {highest_enrollment_borough}")
print(f"Borough with the Lowest Total Student Enrollment: {lowest_enrollment_borough}")
```

Q14: Modify the pivot table from the step above to also reflect the average student body size (*Student Enrollment*) across boroughs, in the same pivot table.

```python
#Q14
pivot_table_ent = sat.groupby('Borough')['Student Enrollment'].agg(['sum', 'mean'])          #Pivot table to summarize total and average student enrollment across all 5 boroughs
pivot_table_ent.columns = ['Total Student Enrollment', 'Average Student Enrollment']
print("Modified Pivot Table - Total and Average Student Enrollment by Borough:")
print(pivot_table_ent)
```

Q15: Create a pivot table of high schools from the Queens borough (*Borough*), containing the SAT section scores (*SAT Section*, *Score*) as columns and the school name (*School Name*) as index. Sort the table in descending order by math section scores.

```python
#Q15
queens_high_sch = sat[(sat['Borough'] == 'Queens') & (sat['School Name'].notnull())]       #Filter data for high schools in Queens

pivot_table_queens = queens_high_sch.pivot_table(values = 'Score', index = 'School Name', columns = 'SAT Section', aggfunc = 'mean')    #Pivot table with SAT section scores as columns

pivot_table_queens_sorted = pivot_table_queens.sort_values(by='Math', ascending=False)        #Sort the table in order by math section scores

print("Pivot Table of High Schools in Queens with SAT Section Scores:")
print(pivot_table_queens_sorted)
```
